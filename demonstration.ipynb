Run Agent 1

```python
from Maze import Maze
import matplotlib.pyplot as plt
from AgentOne import AgentOne
new_maze_1 = Maze(fire_modifier=0.20, mazeSize=75)
fig =plt.figure(figsize=(15,15))
plt.rc('grid', linestyle="-", color='gray', linewidth=2)
grid = plt.gca()
grid.axes.xaxis.set_ticklabels([])
grid.axes.yaxis.set_ticklabels([])
grid.set_ylim(ymin=new_maze_1.mazeSize-0.5)
grid.set_xlim(xmin=-0.5)
plt.grid(True)
agent_one = AgentOne(new_maze_1)
agent_one.move()
new_maze_1.plot_grid()
```
